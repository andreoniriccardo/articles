## Get the Data

```python
url = "https://www.behindthename.com/namesakes/list/tolkien/alpha"
# Get the contents of the webpage in text format
data  = requests.get(url).text

soup = BeautifulSoup(data,"html.parser")
table = soup.find('table', class_='r1r2')
```

```python
table = soup.find('table', class_='r1r2')
```

```python
# Create empty DF
data = pd.DataFrame(columns=["Name", "Gender", "Race","Other Names"])
i=0
# Loop trough the rows of the HTML table. HTML tables rows have the attribute "tr"
for row in table.find_all("tr"):
    # Loop trough the columns of each row. HTML columns have the attribute "tc"
    col = row.find_all("td")
    if (col != []) and i != 0:
        name = col[0].text
        gender = col[1].text
        race = col[2].text
        other_names = col[3].text
        data = pd.concat([data, pd.DataFrame.from_dict({"Name":[name], "Gender":[gender], "Race":[race],"Other Names":[other_names]})])
    
    i+=1
```

```python
data.iloc[::4]
```

```python
lotr_names = list(data.Name.values)
lotr_names = '\n'.join(lotr_names)
lotr_names = lotr_names.lower()

replace_dict = {'a':['â','ä','á'],
                'e':['ê','ë','é'],
                'i':['î','í'],
                'o':['ô','ö','ó'],
                'u':['ú','û',],
                ' ':['-']
}
for new_char in replace_dict.keys():
    for old_char in replace_dict[new_char]:
        lotr_names = lotr_names.replace(old_char, new_char)

print(lotr_names)
```

```python
chars = list(set(lotr_names))
print(chars)
print(len(chars))
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Middle-earth_characters"
# Get the contents of the webpage in text format
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html.parser")
```

```python
soup
```

```python
names = soup.find_all('li')
```

```python
names
```

```python
names_2 = []
for i in names:
    if """<li><a href="/wiki/""" in str(i):
        names_2.append(str(i))
```

```python
names_2
```

```python
lotr_names = []
for i in names_2:
    candidate_name = re.findall(r'"(.*?)"', i)[1]
    conditions = 'Category:' not in candidate_name and 'Middle-earth ' not in candidate_name and 'Tolkien' not in candidate_name and 'Lord of the Rings' not in candidate_name and 'The Hobbit' not in candidate_name
    if conditions:
        lotr_names.append(candidate_name.replace(' (Middle-earth)',''))
```
