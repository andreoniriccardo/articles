```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

test_dimensions = [2, 100000]
print('Average distance between 2 points in a unit hypercube')
for n_dim in test_dimensions:
    distances = []
    for i in range(100000):
        rand_point_1 = np.random.uniform(size=n_dim)
        rand_point_2 = np.random.uniform(size=n_dim)
        dist = np.linalg.norm(rand_point_1-rand_point_2)
        distances.append(dist)
    avg_distance = np.mean(distances)
    print('{} dimensions: {}'.format(n_dim, avg_distance))
```

```python
test_dimensions = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768]
avg_distances = []
for n_dim in test_dimensions:
    print(n_dim)
    distances = []
    for i in range(100000):
        rand_point_1 = np.random.uniform(size=n_dim)
        rand_point_2 = np.random.uniform(size=n_dim)
        dist = np.linalg.norm(rand_point_1-rand_point_2)
        distances.append(dist)
    avg_distances.append(np.mean(distances))
```

```python
fig, ax = plt.subplots()
ax.plot(test_dimensions, avg_distances, linewidth=2, color=sns.color_palette("hls",24)[15], alpha=.9)
plt.xlabel('Number of Dimensions')
plt.title('Average Distance of Random Points in an Unit Hypercube')
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.xaxis.set_ticks_position('none') 
ax.yaxis.set_ticks_position('none')
plt.savefig('avg-dist-rnd-pts.png')
```
